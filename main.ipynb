```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from os import path

data_directory = "Data"
```

# Word Distances

```python
word_distances = np.loadtxt(path.join(data_directory, "word_distances.csv")).astype(np.int32)
```

```python
sns.histplot(x=word_distances, binwidth=1)
plt.xlabel("Edit Distance")
plt.title("Count of Words by Edit Distance")
plt.show()

sns.kdeplot(x=word_distances, bw_adjust=3)
plt.xlabel("Edit Distance")
plt.title("Kernel Density Estimate of Words by Edit Distance")
plt.show()

sns.stripplot(x=word_distances)
plt.xlabel("Edit Distance")
plt.title("Words by Edit Distance")
plt.show()
```

# Wiktionary Cluster Distribution

Note: clusters of size 1 were omitted for these calculations

```python
wikt_distribution = np.loadtxt(path.join(data_directory, "wikt_distribution.csv"))
```

```python
sns.histplot(x=wikt_distribution, binwidth=1)
plt.xlabel("Cluster Size")
plt.title("Count of Clusters by Cluster Size")
plt.show()

sns.kdeplot(x=wikt_distribution, bw_adjust=3)
plt.xlabel("Cluster Size")
plt.title("Kernel Density Estimate of Clusters by Cluster Size")
plt.show()

sns.histplot(x=wikt_distribution, binwidth=1, log_scale=(False, True))
plt.xlabel("Cluster Size")
plt.title("Count of Clusters by Cluster Size")
plt.show()

sns.kdeplot(x=wikt_distribution, bw_adjust=3.5, log_scale=(False, True))
plt.xlabel("Cluster Size")
plt.title("Kernel Density Estimate of Clusters by Cluster Size")
plt.show()

sns.stripplot(x=wikt_distribution)
plt.xlabel("Cluster Size")
plt.title("Clusters by Cluster Size")
plt.show()
```
